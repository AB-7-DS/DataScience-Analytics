# Task 1: Exploring and Visualizing the Iris Dataset

## üîç Introduction
In this task, we perform basic data exploration and visualization on the famous Iris dataset. The objective is to gain insights into data structure, distribution, and relationships between features.



```python
# Step 1: Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```


## üìÅ Dataset Overview
The Iris dataset contains 150 samples of iris flowers from three species: Setosa, Versicolor, and Virginica. Each sample includes:
- Sepal length and width
- Petal length and width

```python

# Step 2: Load the dataset
iris = sns.load_dataset('iris')  # No need for CSV if using seaborn's built-in dataset
```

## üßÆ Dataset Summary
We examine the structure using `.shape`, `.columns`, and `.head()`, and explore summary statistics using `.describe()`.

```python

# Step 3: Summarize the dataset
print("Shape of dataset:", iris.shape)
print("\nColumn names:", iris.columns.tolist())
print("\nFirst 5 rows:\n", iris.head())
```


## üìä Visualizations
- **Pairplot**: To explore pairwise relationships between variables.
- **Histogram**: To view distribution of each feature.
- **Box plot**: To detect outliers and understand spread.

```python

# Step 4: Scatter Plot - pairplot to see relationships
sns.pairplot(iris, hue='species', corner=True)
plt.suptitle("Scatter Plot Matrix (Pairplot)", y=1.02)
plt.show()

# Step 5: Histogram for each numerical feature
iris.hist(figsize=(10, 8), bins=15, edgecolor='black')
plt.suptitle("Histograms of Iris Features")
plt.tight_layout()
plt.show()

# Step 6: Box Plots to detect outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris, orient="h")
plt.title("Box Plot of Iris Features")
plt.show()
```
